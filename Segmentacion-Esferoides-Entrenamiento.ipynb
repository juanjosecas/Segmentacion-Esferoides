### üìä **Bloque 3 ‚Äì Visualizaci√≥n de m√©tricas de entrenamiento**

Este bloque permite inspeccionar gr√°ficamente el desempe√±o de cada modelo durante el entrenamiento.

* Se define la funci√≥n `graficar_metricas_entrenamiento()`, que recibe como entrada la carpeta del experimento (`runs/segment/...`) y genera gr√°ficos autom√°ticos de:

  * **P√©rdidas** (box loss y segmentation loss en entrenamiento y validaci√≥n).
  * **M√©tricas de evaluaci√≥n** (precision, recall, mAP\@50 y mAP\@50-95).
* Los gr√°ficos se muestran por pantalla y tambi√©n se pueden guardar como archivos `.png` en la misma carpeta del experimento.
* Se incorporan controles de errores: la funci√≥n valida que el archivo `results.csv` exista y maneja faltantes de columnas de manera robusta.
* Opcionalmente, la funci√≥n puede devolver el DataFrame con los resultados por √©poca para facilitar an√°lisis adicionales.

Este bloque facilita la **evaluaci√≥n visual** de la convergencia y el rendimiento de los modelos, ayudando a detectar problemas como sobreajuste, poca mejora, o augmentations excesivos.



```python
# =============================================# BLOQUE 3 ‚Äì Visualizaci√≥n de m√©tricas de entrenamiento YOLO# =============================================import osimport pandas as pdimport matplotlib.pyplot as pltimport seaborn as snsdef graficar_metricas_entrenamiento(folder_path, save_plots=True, return_df=False):    """    Carga los resultados de entrenamiento de YOLOv8/v11 y genera gr√°ficos de m√©tricas.    Par√°metros:    - folder_path (str): ruta a la carpeta del experimento (debe contener results.csv).    - save_plots (bool): si True, guarda los PNGs de los gr√°ficos.    - return_df (bool): si True, devuelve el DataFrame con resultados.    Retorna:    - df (pd.DataFrame): resultados si return_df=True; None si False.    """    # OPTIMIZACI√ìN POTENCIAL: Usar pathlib.Path de forma consistente    # en lugar de os.path para mejor portabilidad y legibilidad.        if not os.path.isdir(folder_path):        print(f"‚ùå La carpeta no existe: {folder_path}")        return None    csv_path = os.path.join(folder_path, "results.csv")    if not os.path.isfile(csv_path):        print(f"‚ùå No se encontr√≥ el archivo 'results.csv' en: {folder_path}")        return None    # Leer el CSV    # OPTIMIZACI√ìN POTENCIAL: A√±adir manejo de excepciones para CSVs corruptos    # o con formato incorrecto, proporcionando mensajes de error m√°s informativos.    df = pd.read_csv(csv_path)    df.columns = df.columns.str.strip()    # üîπ Gr√°fico de p√©rdidas    plt.figure(figsize=(10, 6))    hay_perdidas = False    if "train/box_loss" in df:        plt.plot(df["epoch"], df["train/box_loss"], label="Box loss (train)")        hay_perdidas = True    if "val/box_loss" in df:        plt.plot(df["epoch"], df["val/box_loss"], label="Box loss (val)")        hay_perdidas = True    if "train/seg_loss" in df:        plt.plot(df["epoch"], df["train/seg_loss"], label="Seg loss (train)")        hay_perdidas = True    if "val/seg_loss" in df:        plt.plot(df["epoch"], df["val/seg_loss"], label="Seg loss (val)")        hay_perdidas = True    if hay_perdidas:        plt.title("Curva de p√©rdidas por √©poca", fontsize=14)        plt.xlabel("√âpoca")        plt.ylabel("Loss")        plt.legend()        plt.grid(True)        if save_plots:            plt.savefig(os.path.join(folder_path, "losses.png"), dpi=300, bbox_inches="tight")        plt.show()    else:        print("‚ö†Ô∏è No se encontraron columnas de p√©rdida para graficar.")    # üîπ Gr√°fico de m√©tricas    metricas = {        "metrics/precision(M)": "Precision",        "metrics/recall(M)": "Recall",        "metrics/mAP50(M)": "mAP50",        "metrics/mAP50-95(M)": "mAP50-95"    }    plt.figure(figsize=(10, 6))    hay_metricas = False    for col, label in metricas.items():        if col in df.columns:            plt.plot(df["epoch"], df[col], label=label)            hay_metricas = True    if hay_metricas:        plt.title("M√©tricas de evaluaci√≥n por √©poca", fontsize=14)        plt.xlabel("√âpoca")        plt.ylabel("Valor")        plt.ylim(0, 1.05)        plt.legend()        plt.grid(True)        if save_plots:            plt.savefig(os.path.join(folder_path, "metrics.png"), dpi=300, bbox_inches="tight")        plt.show()    else:        print("‚ö†Ô∏è No se encontraron columnas de m√©tricas para graficar.")    print(f"‚úÖ Gr√°ficos generados para: {folder_path}")    return df if return_df else None
```

### üìë **Bloque 4 ‚Äì Informe automatizado de entrenamientos**

Este bloque genera un **resumen cuantitativo** de todos los modelos entrenados, consolidando sus m√©tricas principales en una √∫nica tabla.

* Se define la funci√≥n `generar_informe_entrenamientos()` que:

  * Recorre todas las carpetas de entrenamiento generadas por YOLO (`runs/segment/...`).
  * Extrae los valores finales de cada m√©trica desde el archivo `results.csv`:

    * `mAP50`, `mAP50-95`, `precision`, `recall`, `val/box_loss`, `val/seg_loss`.
  * Crea un DataFrame resumen con una fila por modelo entrenado.
  * Guarda autom√°ticamente el resultado como archivo `.csv` o `.xlsx`.

* Esta tabla permite:

  * Comparar modelos f√°cilmente.
  * Identificar el mejor modelo seg√∫n la m√©trica deseada.
  * Documentar los resultados de forma clara para informes t√©cnicos o publicaciones.

El informe es reproducible y puede extenderse f√°cilmente para incluir tiempos de entrenamiento, hiperpar√°metros clave, o rendimiento en datos externos.


```python
# =============================================# BLOQUE 4 ‚Äì Informe automatizado de entrenamientos YOLO# =============================================import osimport pandas as pdfrom pathlib import Pathdef generar_informe_entrenamientos(root_runs_dir, modelo_list=None, save_path=None):    """    Recorre las carpetas de entrenamiento y genera un resumen por modelo.    Par√°metros:    - root_runs_dir (str): Ruta al directorio base (ej: "/kaggle/working/runs/segment/")    - modelo_list (list): Lista de nombres de modelo (carpetas dentro de root_runs_dir)                          Si es None, se usa todo lo que haya en el directorio.    - save_path (str): Ruta para guardar el CSV generado (si es None, lo guarda en root_runs_dir)    Retorna:    - df_resultados (pd.DataFrame): Tabla resumen con las m√©tricas principales.    """    # CORRECCI√ìN: Uso consistente de pathlib.Path para manejo de rutas    root = Path(root_runs_dir)    if not root.exists():        raise FileNotFoundError(f"[ERROR] No existe la carpeta {root_runs_dir}")    if modelo_list is None:        modelo_list = [f.name for f in root.iterdir() if f.is_dir()]    resultados = []    # OPTIMIZACI√ìN POTENCIAL: Paralelizar el procesamiento de m√∫ltiples modelos    # usando concurrent.futures.ThreadPoolExecutor para mejorar el rendimiento    # cuando se tienen muchos experimentos.    for modelo in modelo_list:        results_csv = root / modelo / "results.csv"        if not results_csv.exists():            print(f"‚ö†Ô∏è No se encontr√≥ results.csv para el modelo {modelo}")            continue        # OPTIMIZACI√ìN POTENCIAL: A√±adir validaci√≥n de que el CSV no est√© vac√≠o        # y tenga las columnas esperadas antes de procesar.        try:            df = pd.read_csv(results_csv)        except Exception as e:            print(f"‚ö†Ô∏è Error al leer results.csv para {modelo}: {e}")            continue                    if df.empty:            print(f"‚ö†Ô∏è El archivo results.csv para {modelo} est√° vac√≠o")            continue                    last_row = df.iloc[-1]  # √öltima √©poca        # CORRECCI√ìN: Validaci√≥n m√°s robusta del campo 'epoch'        epoch_value = last_row.get("epoch", None)        if epoch_value is None or pd.isna(epoch_value):            epoch_final = len(df) - 1        else:            epoch_final = int(epoch_value)        datos = {            "Modelo": modelo,            "Epoch final": epoch_final,            "mAP50": round(last_row.get("metrics/mAP50(M)", 0.0), 4),            "mAP50-95": round(last_row.get("metrics/mAP50-95(M)", 0.0), 4),            "Precision": round(last_row.get("metrics/precision(M)", 0.0), 4),            "Recall": round(last_row.get("metrics/recall(M)", 0.0), 4),            "Box Loss (val)": round(last_row.get("val/box_loss", 0.0), 4),            "Seg Loss (val)": round(last_row.get("val/seg_loss", 0.0), 4)        }        resultados.append(datos)    df_resultados = pd.DataFrame(resultados)    if save_path is None:        save_path = root / "informe_resumen_entrenamientos.csv"    else:        save_path = Path(save_path)    df_resultados.to_csv(save_path, index=False)    print(f"‚úÖ Informe guardado en: {save_path}")    return df_resultados
```

```python
# Generar resumen desde las carpetas creadas por YOLO
resumen = generar_informe_entrenamientos(
    root_runs_dir="/kaggle/working/runs/segment"
)

# Mostrar el top 3 modelos por mAP50
resumen.sort_values("mAP50", ascending=False).head(3)
```
